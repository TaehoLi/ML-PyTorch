## 2. Prepare learning data

```python
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

import pandas as pd

import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# data call
dat = pd.read_csv('./data/weather_data.csv', skiprows=[0, 1, 2, 3, 4, 5], encoding="cp949")
dat
```

```python
# Extracting and visualizing average temperature values
temp = dat['평균기온(℃)']

temp.plot()
plt.show()
```

```python
# data partitoning
train_x = temp[:1461] # 2011.1.1 ~ 2014.12.31
test_x = temp[1461:]  # 2015.1.1 ~ 2016.12.31

# change to NUMPY array
train_x = np.array(train_x)
test_x = np.array(test_x)
```

```python
# size of partial series (input)
ATTR_SIZE = 180   # 6 months

tmp = []
train_X = []
# Slides window one by one, extracting train data
for i in range(0, len(train_x) - ATTR_SIZE):
    tmp.append(train_x[i:i+ATTR_SIZE])
train_X = np.array(tmp)

# Convert train data into a data frame and print
pd.DataFrame(train_X)
```

## 3. Build NN

```python
# NN
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(180, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 128)
        self.fc4 = nn.Linear(128,180)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        x = F.relu(self.fc3(x))
        x = self.fc4(x)
        return x
    
# instance
model = Net()
```

## 4. Model fitting

```python
# loss func
criterion = nn.MSELoss()

# optimization
optimizer = optim.Adam(model.parameters(), lr=0.001)

# fitting
for epoch in range(1000):
    total_loss = 0
    d = []
    
    # train data -> minibatch
    for i in range(100):
        # indexing
        index = np.random.randint(0, 1281)
        # split minibatch
        d.append(train_X[index])
    
    d = np.array(d, dtype='float32')
    # Calc Graph Configuration
    d = Variable(torch.from_numpy(d))
    
    # Slope Initialization
    optimizer.zero_grad()
    # calc FP
    output = model(d)
    # calc loss
    loss = criterion(output, d)
    # calc BP
    loss.backward()
    # update weights
    optimizer.step()
    # Calc cumulative losses
    total_loss += loss.item()
        
    # print cum_loss every 100 epochs
    if (epoch+1) % 100 == 0:
        print(epoch+1, total_loss)
```

```python
# data plotting
plt.plot(d.data[0].numpy(), label='original')
plt.plot(output.data[0].numpy(), label='output')
plt.legend(loc='upper right')
plt.show()
```

## 5. Calculate anomaly score

```python
tmp = []
test_X = []

# Divide test data into six unit(month)
tmp.append(test_x[0:180])
tmp.append(test_x[180:360])
tmp.append(test_x[360:540])
tmp.append(test_x[540:720])
test_X = np.array(tmp, dtype="float32")

pd.DataFrame(test_X)
```

```python
# model fit
d = Variable(torch.from_numpy(test_X))
output = model(d)

# data plotting
plt.plot(test_X.flatten(), label='original')
plt.plot(output.data.numpy().flatten(), label='prediction')
plt.legend(loc='upper right')
plt.show()
```

```python
# Calculate anomaly score
test = test_X.flatten()
pred = output.data.numpy().flatten()

total_score = []
for i in range(0, 720):
    dist = (test[i] - pred[i])
    score = pow(dist, 2)
    total_score.append(score)
    
# normalization
total_score = np.array(total_score)
max_score = np.max(total_score)
total_score = total_score / max_score

print(total_score[-10:])
```
